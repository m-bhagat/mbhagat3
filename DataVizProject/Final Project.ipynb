## General Information <a class="anchor" id="first-bullet"></a>


**What is the "name" of the dataset?**

Loan data from Lending Club<br>

**Summary**<br>

This data set represents thousands of loans made through the Lending Club platform, which is a platform that allows individuals to lend to other individuals. Of course, not all loans are created equal. Someone who is a essentially a sure bet to pay back a loan will have an easier time getting a loan with a low interest rate than someone who appears to be riskier. And for people who are very risky? They may not even get a loan offer, or they may not have accepted the loan offer due to a high interest rate. It is important to keep that last part in mind, since this data set only represents loans actually made, i.e. do not mistake this data for loan applications! ## Preliminary Dataset Insights <a class="anchor" id="third-bullet"></a>

- `months_since_last_delinq` represents months elapsed since a person has missed their repayment. Hence we assume that the `null` values represent people that haven't missed their payment and replace it with 0. This gives us another column with no `null` values.
- Replaced `Source Verified` with `Verified` in the `verified_income` column as we assumed it to be a system nuance rather than any unique category
- We tried to see if there was an overlap between `null` values in `emp_title` and `emp_length`, which could mean that the person is unemployed. But since there was no overlap between the two, this assumption was wrong
